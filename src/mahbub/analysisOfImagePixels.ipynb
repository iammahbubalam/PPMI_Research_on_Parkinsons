```python
import random
import os
import statistics
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def calculate_pixel_stats(image_path):
    # Open the image
    img = Image.open(image_path)
    
    # Convert image to grayscale for simplicity
    img = img.convert('L')
    
    # Get pixel data
    pixels = list(img.getdata())
    
    # Calculate lowest and highest pixel values
    lowest_pixel = min(pixels)
    highest_pixel = max(pixels)
    
    # Calculate mean, mode, median, and RMS pixel values
    mean_pixel = np.mean(pixels)
    mode_pixel = statistics.mode(pixels)
    median_pixel = np.median(pixels)
    rms_pixel = np.sqrt(np.mean(np.square(pixels)))
    
    return lowest_pixel, highest_pixel, mean_pixel, mode_pixel, median_pixel, rms_pixel

# Folder path containing PNG images
folder_path = r"B:\Sami Sir Research\Data set\pngimage\hc"

# Calculate pixel stats for each image
stats = []
lowest_values = []
highest_values = []
mean_values = []
mode_values = []
median_values = []
rms_values = []

all_files = os.listdir(folder_path)

# Calculate 10% of the total number of files
num_files_to_select = int(1 * len(all_files))

# Randomly select 10% of the files
selected_files = random.sample(all_files, num_files_to_select)

# for filename in os.listdir(folder_path):
for filename in selected_files:
    if filename.endswith(".png"):
        image_path = os.path.join(folder_path, filename)
        pixel_stats = calculate_pixel_stats(image_path)
        stats.append((filename, pixel_stats))
        lowest_values.append(pixel_stats[0])
        highest_values.append(pixel_stats[1])
        mean_values.append(pixel_stats[2])
        mode_values.append(pixel_stats[3])
        median_values.append(pixel_stats[4])
        rms_values.append(pixel_stats[5])

# Helper function to plot histogram
def plot_histogram(values, title):
    plt.figure(figsize=(8, 5))
    plt.hist(values, bins=20, color='blue', alpha=0.7)
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.grid(True)
    plt.show()

# Helper function to plot box plot
def plot_box_plot(values, title):
    plt.figure(figsize=(8, 5))
    plt.boxplot(values, vert=False)
    plt.xlabel('Pixel Value')
    plt.title(title)
    plt.grid(True)
    plt.show()

# Helper function to plot scatter plot
def plot_scatter_plot(x_values, y_values, title):
    plt.figure(figsize=(8, 5))
    plt.scatter(x_values, y_values, color='red', alpha=0.5)
    plt.xlabel('Image Index')
    plt.ylabel('Pixel Value')
    plt.title(title)
    plt.grid(True)
    plt.show()

# Helper function to plot line plot
def plot_line_plot(values, title):
    plt.figure(figsize=(8, 5))
    plt.plot(values, marker='o', color='green')
    plt.xlabel('Image Index')
    plt.ylabel('Pixel Value')
    plt.title(title)
    plt.grid(True)
    plt.show()

# Helper function to plot heatmap
def plot_heatmap(image_path):
    img = Image.open(image_path)
    img = img.convert('L')
    pixels = np.array(img)
    plt.figure(figsize=(8, 6))
    plt.imshow(pixels, cmap='hot', interpolation='nearest')
    plt.colorbar()
    plt.title('Heatmap of Pixel Values')
    plt.show()

# Plot histograms
# plot_histogram(lowest_values, 'Histogram of Lowest Pixel Values')
# plot_histogram(highest_values, 'Histogram of Highest Pixel Values')
plot_histogram(mean_values, 'Histogram of Mean Pixel Values')
plot_histogram(mode_values, 'Histogram of Mode Pixel Values')
plot_histogram(median_values, 'Histogram of Median Pixel Values')
plot_histogram(rms_values, 'Histogram of RMS Pixel Values')

# Plot box plots
# plot_box_plot(lowest_values, 'Box Plot of Lowest Pixel Values')
# plot_box_plot(highest_values, 'Box Plot of Highest Pixel Values')
plot_box_plot(mean_values, 'Box Plot of Mean Pixel Values')
plot_box_plot(mode_values, 'Box Plot of Mode Pixel Values')
plot_box_plot(median_values, 'Box Plot of Median Pixel Values')
plot_box_plot(rms_values, 'Box Plot of RMS Pixel Values')

# Generate random indices for scatter plot
indices = np.arange(len(lowest_values))

# Plot scatter plots
# plot_scatter_plot(indices, lowest_values, 'Scatter Plot of Lowest Pixel Values')
# plot_scatter_plot(indices, highest_values, 'Scatter Plot of Highest Pixel Values')
plot_scatter_plot(indices, mean_values, 'Scatter Plot of Mean Pixel Values')
plot_scatter_plot(indices, mode_values, 'Scatter Plot of Mode Pixel Values')
plot_scatter_plot(indices, median_values, 'Scatter Plot of Median Pixel Values')
plot_scatter_plot(indices, rms_values, 'Scatter Plot of RMS Pixel Values')

# Plot line plots
# plot_line_plot(lowest_values, 'Line Plot of Lowest Pixel Values')
# plot_line_plot(highest_values, 'Line Plot of Highest Pixel Values')
plot_line_plot(mean_values, 'Line Plot of Mean Pixel Values')
plot_line_plot(mode_values, 'Line Plot of Mode Pixel Values')
plot_line_plot(median_values, 'Line Plot of Median Pixel Values')
plot_line_plot(rms_values, 'Line Plot of RMS Pixel Values')

# Plot heatmaps (for each image)
# for filename, _ in stats:
#     image_path = os.path.join(folder_path, filename)
#     plot_heatmap(image_path)
```

```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Function to calculate standard deviation of pixel values in an image
def calculate_std_dev(image_path):
    img = Image.open(image_path)
    img = img.convert('L')  # Convert to grayscale
    pixels = np.array(img).flatten()  # Flatten the pixel array
    std_dev = np.std(pixels)
    return std_dev


def standard_dev_plot(path_to_the_folder, n):
    # Get a list of all image files in the folder
    image_files = [f for f in os.listdir(path_to_the_folder) if f.endswith('.png')]
    
    # Calculate 10% of the total number of images
    num_images_to_select = int(n * len(image_files))
    
    # Randomly select 10% of the images
    selected_images = random.sample(image_files, num_images_to_select)
    
    # List to store standard deviation values 
    std_dev_values = []
    
    # Calculate standard deviation for each selected image
    for filename in selected_images:
        image_path = os.path.join(path_to_the_folder, filename)
        std_dev = calculate_std_dev(image_path)  # Corrected line
        std_dev_values.append(std_dev)
        
    plot_line_plot(std_dev_values, 'Line Plot of Standard Deviation of Pixel Values')

    # Plot standard deviation values
    plt.figure(figsize=(8, 6))
    plt.hist(std_dev_values, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
    plt.xlabel('Standard Deviation')
    plt.ylabel('Frequency')
    plt.title('Distribution of Standard Deviation of Pixel Values')
    plt.grid(True)
    plt.show()
    indices = np.arange(len(std_dev_values))
    plot_scatter_plot(indices, std_dev_values, 'Scatter Plot of Standard Deviation of Pixel Values')

     
folder_path_hc = r"B:\Sami Sir Research\Data set\pngimage\hc"
folder_path_pd = r"B:\Sami Sir Research\Data set\pngimage\pd"
standard_dev_plot(folder_path_hc,1)
standard_dev_plot(folder_path_pd,.1)
```

```python
def calculate_pdf(image_path, num_bins=256):
    # Open the image
    img = Image.open(image_path)
    
    # Convert image to grayscale for simplicity
    img = img.convert('L')
    
    # Get pixel data
    pixels = np.array(img).flatten()
    
    # Calculate histogram
    hist, bins = np.histogram(pixels, bins=num_bins, range=(0, 256), density=True)
    
    # Calculate bin centers
    bin_centers = (bins[:-1] + bins[1:]) / 2
    
    return bin_centers, hist
img_path = r"B:\Sami Sir Research\Data set\pngimage\pd\3077_2014-03-07_14_32_46.0_slice37.png"
bin_centers, pdf = calculate_pdf(img_path)

plt.plot(bin_centers, pdf)
plt.xlabel('Pixel Value')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF)')
plt.grid(True)
plt.show()
```

```python
from collections import Counter
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Function to calculate mean pixel value
def calculate_mean_pixel_value(image_path):
    img = Image.open(image_path)
    img = img.convert('L')
    pixels = np.array(img)
    mean_pixel_value = np.mean(pixels)
    return mean_pixel_value

def calculate_mode_pixel_value(image_path):
    # Open the image
    img = Image.open(image_path)
    # Convert image to grayscale
    img = img.convert('L')
    # Get pixel data
    pixels = list(img.getdata())
    # Count the frequency of each pixel value
    pixel_counts = Counter(pixels)
    # Get the pixel value(s) with the highest frequency
    mode_values = pixel_counts.most_common(1)   
    # Return the mode pixel value
    mode_pixel_value = mode_values[0][0]
    return mode_pixel_value

# Folder path containing PNG images
folder_path = r"B:\Sami Sir Research\Data set\pngimage\hc"

# List to store paths of images with mean pixel value greater than 80
selected_images = []

# Iterate over images in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".png"):
        image_path = os.path.join(folder_path, filename)
        mean_pixel_value = calculate_mode_pixel_value(image_path)
        if mean_pixel_value > 50:
            selected_images.append((filename, image_path))


# Plot selected images
num_images = len(selected_images)
num_rows = (num_images + 3) // 4  # Calculate number of rows needed
plt.figure(figsize=(15, 5 * num_rows))  # Adjust figure size based on number of rows
for i, (filename, image_path) in enumerate(selected_images, 1):
    img = Image.open(image_path)
    plt.subplot(num_rows, 4, i)  # Adjust subplot layout based on number of rows
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    plt.title(filename)  # Set subplot title as the file name

plt.tight_layout()
plt.show()
```
