```python
 import pandas as pd
import numpy as np
import math

# Load the dataset
data = pd.read_csv("B:/Projects/PycharmProjects/PPMI_Research_on_Parkinson's/src/mahbub/processed_ppmi_master_dataset.csv")

# Numerical features
numerical_features = [
    'Area', 'Circularity', 'ConvexArea', 'EquivDiameter', 'Extent',
    'FilledArea', 'Kurtosis', 'Major axis length', 'Mean',
    'Minor axis length', 'PA_ratio', 'Shannon_Entropy', 'Skewness',
    'Solidity', 'Standard Deviation', 'brightness', 'contrast',
    'correlation', 'dissimilarity', 'energy', 'gabor_energy',
    'gabor_entropy', 'homogeneity', 'lbp_energy', 'lbp_entropy'
]

# Round numerical columns to three decimal points
data[numerical_features] = data[numerical_features].round(3)

# Group by 'Visit Class'
grouped_data = data.groupby('Visit Class')

# Dictionary to store patient IDs for each bin edge, grouped by feature and Visit Class
patient_ids_by_bin = {feature: {group_name: [] for group_name, _ in grouped_data} for feature in numerical_features}

# Now iterate over each feature and group as before
for feature in numerical_features:
    print(f"\nFeature: {feature}")
    
    for group_name, group_df in grouped_data:
        print(f"\nGroup: {group_name}")
        
        # Extract feature data and corresponding patient IDs for the current group, dropping NaN values
        feature_data = group_df[feature].dropna()
        patient_ids = group_df.loc[feature_data.index, 'Patient ID']
        
        # Calculate the minimum and maximum values for the feature in the current group
        min_value = math.floor(feature_data.min()) if feature_data.min() > 1 else feature_data.min()
        max_value = math.ceil(feature_data.max()) if feature_data.min() > 1 else feature_data.max()
        num_bins = 5

        # Calculate bin width
        bin_width = (max_value - min_value) / num_bins
        
        # Create bin edges
        bin_edges = [min_value + i * bin_width for i in range(num_bins + 1)]

        # Calculate histogram counts for each bin
        hist_counts, _ = np.histogram(feature_data, bins=bin_edges)

        # Calculate density (frequency normalized by total count)
        density = hist_counts / len(feature_data)

        # Calculate percentage (density as percentage of total)
        percentage = density * 100
        
        # Calculate frequency (raw counts in each bin)
        frequency = hist_counts
        
        # Print the values with three decimal points
        print(f"Min Value: {min_value:.3f}, Max Value: {max_value:.3f}")
        print(f"Bin Edges: {[round(edge, 3) for edge in bin_edges]}")
        print(f"Density: {density}")
        print(f"Percentage: {percentage}")
        print(f"Frequency: {frequency}")

        # Save patient IDs for each bin edge
        for i, edge in enumerate(bin_edges[:-1]):
            bin_patients = patient_ids[(feature_data >= edge) & (feature_data < bin_edges[i + 1])]
            patient_ids_by_bin[feature][group_name].append(list(bin_patients))

# Example of accessing patient IDs for a specific feature, group, and bin edge
example_feature = numerical_features[0]
example_group = list(grouped_data.groups.keys())[0]
example_bin_index = 0
example_bin_edge = patient_ids_by_bin[example_feature][example_group][example_bin_index]

print(f"\nExample: Patient IDs for Feature '{example_feature}', Group '{example_group}', Bin Edge {example_bin_index}:")
print(example_bin_edge)
```

```python
import pandas as pd
from collections import defaultdict

# Load the dataset
data = pd.read_csv("B:/Projects/PycharmProjects/PPMI_Research_on_Parkinson's/src/mahbub/processed_ppmi_master_dataset.csv")

# Group the data by 'Visit Class'
grouped_data = data.groupby('Visit Class')

# Categorical features
categorical_features = [
    'NP3SPCH', 'NP3FACXP', 'NP3RIGN', 'NP3RIGRU', 'NP3RIGLU', 'NP3RIGRL', 
    'NP3RIGLL', 'NP3FTAPR', 'NP3FTAPL', 'NP3HMOVR', 'NP3HMOVL', 'NP3PRSPR', 
    'NP3PRSPL', 'NP3TTAPR', 'NP3TTAPL', 'NP3LGAGR', 'NP3LGAGL', 'NP3RISNG', 
    'NP3GAIT', 'NP3FRZGT', 'NP3PSTBL', 'NP3POSTR', 'NP3BRADY', 'NP3PTRMR', 
    'NP3PTRML', 'NP3KTRMR', 'NP3KTRML', 'NP3RTARU', 'NP3RTALU', 'NP3RTARL', 
    'NP3RTALL', 'NP3RTALJ', 'NP3RTCON', 'DYSKPRES', 'NHY', 'NP2SPCH', 'NP2SALV', 
    'NP2SWAL', 'NP2EAT', 'NP2DRES', 'NP2HYGN', 'NP2HWRT', 'NP2HOBB', 'NP2TURN', 
    'NP2TRMR', 'NP2RISE', 'NP2WALK', 'NP2FREZ', 'NP1SLPN', 'NP1SLPD', 'NP1PAIN', 
    'NP1URIN', 'NP1CNST', 'NP1LTHD', 'NP1FATG', 'NP1COG', 'NP1HALL', 'NP1DPRS', 
    'NP1ANXS', 'NP1APAT', 'NP1DDS', 'CONCOHORT'
]

# Dictionary to store patient IDs for each value, for each group, for each feature
patient_ids_by_value = defaultdict(lambda: defaultdict(lambda: defaultdict(list)))

# Iterate through each group and feature to collect patient IDs
for feature in categorical_features:
    print(f"\nFeature: {feature}")
    for group_name, group_df in grouped_data:
        print(f"\nGroup: {group_name}")
        
        # Initialize a set to collect patient IDs for each value in the current group and feature
        value_patient_ids = defaultdict(list)
        
        # Iterate through each unique value in the current group and feature
        for value in group_df[feature].dropna().unique():
            # Get patient IDs associated with the current value
            patients = group_df.loc[group_df[feature] == value, 'Patient ID'].tolist()
            value_patient_ids[value].extend(patients)
        
        # Store the collected patient IDs in the dictionary
        patient_ids_by_value[feature][group_name] = value_patient_ids
        
        value_counts = group_df[feature].value_counts()
        
        # Calculate density (proportion of occurrences for each value)
        total_count = len(group_df)
        density = value_counts / total_count
        print(f"\nFeature: {feature}")
        for value, count in value_counts.items():
            dens = density[value]
            print(f"Value: {value}, Frequency: {count}, Density: {dens:.4f}")

# Example usage: Print patient IDs for a specific feature, group, and value
example_feature = 'NP1DDS'
example_group = 'V4'
example_value = 1

print(f"Patient IDs for Feature '{example_feature}', Group '{example_group}', Value '{example_value}':")
print(patient_ids_by_value[example_feature][example_group][example_value])
```
