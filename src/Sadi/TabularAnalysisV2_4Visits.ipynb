### MDS-UPDRS_Part_III_04Apr2024

```python
mds_updrs_part_three = pd.read_csv("NecessaryCSVFiles/MDS-UPDRS_Part_III_04Apr2024.csv")

mds_updrs_part_three
```

```python
# Calculate the percentage of missing data in each column
missing_data_percent = mds_updrs_part_three.isnull().mean() * 100

missing_data_percent
```

```python
# Calculate the percentage of missing data in each column
missing_data_percent = mds_updrs_part_three.isnull().mean() * 100

# Create a mask for columns that have more than 10% missing data
mask = missing_data_percent > 10

# Drop these columns from the DataFrame
mds_updrs_part_three = mds_updrs_part_three.loc[:, ~mask]

mds_updrs_part_three
```

```python
mds_updrs_part_three = mds_updrs_part_three.drop(['REC_ID', 'PAG_NAME', 'INFODT','PDTRTMNT', 'DBSYN', 'EXAMDT', 'ORIG_ENTRY', 'LAST_UPDATE'], axis=1)
```

```python
# mds_updrs_part_three.columns
```

```python
# Define the columns for which fill will be mode
columns_to_fill = ['NP3SPCH', 'NP3FACXP', 'NP3RIGN', 'NP3RIGRU', 'NP3RIGLU', 'NP3RIGRL', 'NP3RIGLL', 'NP3FTAPR', 'NP3FTAPL', 'NP3HMOVR', 'NP3HMOVL', 'NP3PRSPR', 'NP3PRSPL', 'NP3TTAPR', 'NP3TTAPL', 'NP3LGAGR', 'NP3LGAGL', 'NP3RISNG', 'NP3GAIT', 'NP3FRZGT', 'NP3PSTBL', 'NP3POSTR', 'NP3BRADY', 'NP3PTRMR', 'NP3PTRML', 'NP3KTRMR', 'NP3KTRML', 'NP3RTARU', 'NP3RTALU', 'NP3RTARL', 'NP3RTALL', 'NP3RTALJ', 'NP3RTCON', 'DYSKPRES', 'NHY']

# Fill missing values and values equal to 101 with mode for specified columns
for col in columns_to_fill:
    mode_value = mds_updrs_part_three[col].mode()[0]
    mds_updrs_part_three[col] = mds_updrs_part_three[col].replace(101, np.nan)  # Replace 101 with NaN
    mds_updrs_part_three[col] = mds_updrs_part_three[col].fillna(mode_value)  # Fill NaN with mode
```

```python
mds_updrs_part_three
```

```python
# Group by 'PATNO' and filter to keep only those groups where all five event IDs are present
mds_updrs_part_three = mds_updrs_part_three.loc[mds_updrs_part_three['EVENT_ID'].isin(['BL', 'V04', 'V06', 'V08'])]

mds_updrs_part_three = mds_updrs_part_three.groupby('PATNO').filter(lambda x: set(['BL', 'V04', 'V06', 'V08']).issubset(set(x['EVENT_ID'])))
```

```python
mds_updrs_part_three = mds_updrs_part_three.groupby(['PATNO', 'EVENT_ID']).max().reset_index()
```

```python
mds_updrs_part_three_copy = mds_updrs_part_three.copy()

mds_updrs_part_three_copy
```
