## Motor Data No Duplicate

```python
# Load the DataFrame
mds_updrs_merged_p1_p1p_p2p_p3 = pd.read_csv('CreatingDataset/FourVisits/mds_updrs_merged_p1_p1p_p2p_p3.csv')

# Convert 'INFODT' to datetime format
mds_updrs_merged_p1_p1p_p2p_p3['INFODT'] = pd.to_datetime(mds_updrs_merged_p1_p1p_p2p_p3['INFODT'], format='%m/%Y')

# Identify numeric columns
numeric_cols = mds_updrs_merged_p1_p1p_p2p_p3.select_dtypes(include=[np.number]).columns.tolist()

# Drop 'PATNO' and 'INFODT' from numeric_cols
numeric_cols = [col for col in numeric_cols if col not in ['PATNO', 'INFODT', 'EVENT_ID']]

# Group by 'PATNO' and 'INFODT', and take the maximum value for numeric columns
mds_updrs_merged_p1_p1p_p2p_p3_no_duplicate = mds_updrs_merged_p1_p1p_p2p_p3.groupby(['PATNO', 'INFODT', 'EVENT_ID'])[numeric_cols].max().reset_index()

# Convert 'INFODT' back to 'MM/YYYY' format
mds_updrs_merged_p1_p1p_p2p_p3_no_duplicate['INFODT'] = mds_updrs_merged_p1_p1p_p2p_p3_no_duplicate['INFODT'].dt.strftime('%m/%Y')

# Save the result to a CSV file
mds_updrs_merged_p1_p1p_p2p_p3_no_duplicate.to_csv('CreatingDataset/FourVisits/mds_updrs_merged_p1_p1p_p2p_p3_no_duplicate.csv', index=False)
```
